### Data Details
Features - 32 to be converted to 33
Target feature - 1 Type of road added from DataLabel into both gpsmpuleft and gpsmpuright

```python
RoadType = DataLabels.dirt_road + DataLabels.cobblestone_road + DataLabels.asphalt_road
RoadType[:5]
```

```python
for i in range(len(RoadType)):
    if RoadType[i] == "NoNoYes": RoadType[i] = "Asphalt"
    elif RoadType[i] == "NoYesNo": RoadType[i] = "Cobblestone"
    if RoadType[i] == "YesNoNo": RoadType[i] = "Dirt"
set(RoadType)
```

```python
DataGpsMpuLeft['RoadType'], DataGpsMpuRight['RoadType'] = RoadType, RoadType
# del RoadType
DataGpsMpuLeft.head(2)
```
